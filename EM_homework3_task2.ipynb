```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5, 210)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x, y, color="red")
plt.plot(x, y2, color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
```

```python
#ОТВЕТ:
# Прямые не кажутся перпендикулярными из-за того, что по осям разный масштаб.
# Если установить одинаковые пределы по осям и равное соотношение сторон, то прямые будут выглядеть перпендикулярными.

x = np.linspace(-5, 5, 210)
y = 3*x+1
y2 = (-1/3)*x+1
plt.plot(x, y, color="red")
plt.plot(x, y2, color="blue")
plt.xlabel("x")
plt.ylabel("y")
ax = plt.axes()
ax.set_ylim(-5, 5)
ax.set_aspect('equal')
plt.grid(True)
```
