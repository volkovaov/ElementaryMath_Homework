```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, np.sqrt(25 - x**2))
plt.plot(x, -np.sqrt(25 - x**2))
plt.axis("square")
```

```python
x = np.linspace(-3, 3, 100)
a = 3
b = 2
plt.plot(x, np.sqrt((1 - x**2/a**2)/b**2))
plt.plot(x, -np.sqrt((1 - x**2/a**2)/b**2))
#plt.axis("square")
#ax = plt.axes()
#ax.set_ylim(-3,3)
```

```python
x = np.linspace(-10, 10, 200)
x[x!=0]

plt.plot(x, 1/x)
```

```python
#Чтобыизбежать разрыва (х=0), разбиваем график на две части: выкалываем промежуток (-0,1; 0,1).
x = np.linspace(-3, -0.1, 100)
x2 = np.linspace(0.1, 3, 100)

plt.plot(x, 1/x)
plt.plot(x2, 1/x2)
```
