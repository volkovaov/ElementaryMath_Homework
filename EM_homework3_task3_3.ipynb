```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Перевод полярных координат в декартовы
def polar2decart(r, phi):
    return (r*np.cos(phi), r*np.sin(phi))

x, y = polar2decart(5, 3)
print(x,y)
```

```python
# 2. График окружности в полярных координатах
theta = np.arange(0., 2., 1./180.)*np.pi
plt.polar(theta, [2]*len(theta))
plt.show()
```

```python
# 3. График отрезка прямой в полярных координатах
fig = plt.figure()
ax = fig.add_subplot(111, projection='polar')

phi = [np.pi/4, 3/8*np.pi]
rho = [5, 3]
ax.plot(phi,rho, color = "red")
plt.show()
```
