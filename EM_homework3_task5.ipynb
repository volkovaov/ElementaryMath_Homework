```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 6, 2)
Y = np.arange(-50, 50, 2)
X, Y = np.meshgrid(X, Y)
Z = 2*X + 2*Y + 100
Z2 = 2*X + 2*Y
ax.plot_surface(X, Y, Z, color="green")
ax.plot_surface(X, Y, Z2, color="orange")
show()
```

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-80, 80, 2)
y = np.arange(-50, 50, 2)
x, y = np.meshgrid(x, y)
a=3
b=2
z = (x**2/a**2 - y**2/b**2)/2
ax.plot_surface(x, y, z, color="red")
show()
```
