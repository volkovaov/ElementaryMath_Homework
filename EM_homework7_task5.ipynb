```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Q (x, y, z):
    return(x**2 + y**2 + z**2)
t = np.linspace(-0.3, 0, 201)
plt.plot(t, Q(1.4 + 0.1*t, t, 0.4 + 2.1*t))

plt.xlabel("y")
plt.ylabel("Q")
plt.grid(True)
plt.show()
```
